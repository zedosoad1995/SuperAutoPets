```python
attack = [1, 2, 3, 4]
defense = [1, 2, 3, 4]

stats = []

for att in attack:
    for defen in defense:
        stats.append([att, defen])
```

```python
import copy

def get_animals_permutations(animals):
    ret = []
    
    for perm in permutations(animals):
        ret.append(list(perm))
        
    return ret

def battle(player_, cpu_):
    idx_player = 0
    idx_cpu = 0
    
    player = copy.deepcopy(player_)
    cpu = copy.deepcopy(cpu_)
    
    while True:
        # damage cpu->player
        player[idx_player][1] -= cpu[idx_cpu][0]
        # damage player->cpu
        cpu[idx_cpu][1] -= player[idx_player][0]
        
        if player[idx_player][1] <= 0:
            idx_player += 1
            
        if cpu[idx_cpu][1] <= 0:
            idx_cpu += 1
            
        if idx_cpu == 5 and idx_player == 5:
            return "draw"
        if idx_cpu == 5:
            return "loss"
        if idx_player == 5:
            return "win"
```

```python
from itertools import permutations, combinations

all_animals = []
for comb in combinations(range(0, len(stats)), 5):
    all_animals.append([stats[i] for i in comb])
        
for animals in all_animals:
    animals_perm = get_animals_permutations(animals)
    
    
    best_score = -999
    best_score_details = []
    best_combination = []
    
    for animals_player in animals_perm:
        wins = 0
        losses = 0
        draws = 0
        
        for animals_cpu in animals_perm:
            battle_result = battle(animals_player, animals_cpu)
            
            if battle_result == 'win':
                wins += 1
            elif battle_result == 'loss':
                losses += 1
            else:
                draws += 1
                
        score = wins - losses
        
        if score > best_score:
            best_score = score
            best_score_details = {'wins': wins, 'losses': losses, 'draws': draws}
            best_combination = animals_player
            
    print(best_score, best_score_details, best_combination)
```
